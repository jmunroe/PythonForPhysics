# Physics with Python

We first set up Python to allow for plotting and doing numerical work.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

Bessel functions are special functions and are included in the scipy.special module

```python
from scipy.special import jn, yn

x = np.arange(0, 20, 0.1)
plt.plot(x, jn(0, x))
plt.plot(x, yn(0, x))
```

We can also add basic iteractions, such as varying the frequency

```python
from ipywidgets.widgets import interact

x = np.arange(0, 20, 0.01)

def pltbessel(f=1.0):
    plt.plot(x, jn(0, 2*np.pi*x*f))
    plt.title('Vary $f$ to see the effect of changing frequency.')
    plt.show()
   
interact(pltbessel, f=(0.01,2,0.01))
```

Sometimes we want to understand how the function varies radially.

```python
x = np.arange(-20, 20, .5)
y = np.arange(-20, 20, .5)
X, Y = np.meshgrid(x, y)
R = np.sqrt(X**2+Y**2)
Z = jn(0, R)
plt.contourf(X, Y, Z, cmap=cm.coolwarm, vmin=-1, vmax=+1)
plt.colorbar()
```
